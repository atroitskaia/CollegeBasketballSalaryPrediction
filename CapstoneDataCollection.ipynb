```python
# Combining two data frames and also changing salary column to numeric value

combined_df = pd.concat([salary_df, salary_df1], ignore_index=True)

# Convert Salary to numeric, removing the '$' and ',' characters
combined_df['Salary'] = combined_df['Salary'].replace('[\$,]', '', regex=True).astype(float)

# Sort the combined dataframe by Year and then by Salary
combined_sorted_df = combined_df.sort_values(['Year', 'Salary'], ascending=[True, False])

# We should reset the index
combined_sorted_df.reset_index(drop=True, inplace=True)


# print(df_combined_sorted.to_string(index=True))

# just in case going to save this dataset
combined_sorted_df.to_csv('combined_basketball_salaries.csv', index=False)
```

```python
combined_sorted_df.head()
```

```python
# Reshaping 'combined_sorted_df' from long format to short format
duplicates = combined_sorted_df.duplicated(subset=['Player', 'Year'], keep=False)
if duplicates.any():
    print("Warning: Duplicate entries found:")
    print(combined_sorted_df[duplicates])

    combined_sorted_df = combined_sorted_df.drop_duplicates(subset=['Player', 'Year'], keep='first')

# Reshaping the dataframe
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pivot.html
df_reshaped = combined_sorted_df.pivot(index='Player', columns='Year', values='Salary')

# Reset the index to make 'Player' a column again
df_reshaped.reset_index(inplace=True)

# Fill NaN values with 0
df_reshaped = df_reshaped.fillna(0) 

# Display the result
print(df_reshaped.to_string(index=False))
```

```python
df_reshaped.head()
```

```python
# After reshaping the data frame, it makes it look a little crazy. 
# I didn't think about players that werent playing in 2009 or after certain years. 
# I think my next step is to forget the idea of reshaping the data set and just calcualte the average for each player 
# insted of looking at the player's salary year by year. 
```

```python
# Let's try again to make sense out of it 
def average_salary(row):
    salary_col = row.index[row.index != 'Player']
    salaries = row[salary_col].dropna()
    if len(salaries) > 0:
        return salaries.mean()
    else:
        return np.nan
# Calculating average salary for each player
df_reshaped['Average Salary'] = df_reshaped.apply(average_salary, axis=1)

# Droping all the columns except 'Player' and 'Average Salary'
average_salary_df = df_reshaped[['Player', 'Average Salary']]
average_salary_df['Average Salary'].astype(np.int64)
# pd.options.display.float_format = '{:.2f}'.format

print(average_salary_df.to_string(index=False))
```
